```python
import librosa
import librosa.display
import IPython.display as ipd
import numpy as np 
import pandas as pd
import pickle
from collections import defaultdict
%matplotlib inline
```

# FEATURE EXTRACTION

```python
files = []
#update range to extract features in batches - 5,000 at a time 
for i in range(10000,14999):
    #filenames in the format: 'train1.wav'
    new_file = 'train' + str(i) + '.wav'
    files.append(new_file)
```

```python
df = pd.DataFrame()
for file in files:
    file_dict = defaultdict()
    X, sample_rate = librosa.load(file)
    stft = np.abs(librosa.stft(X))
    mfccs = np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T,axis=0)
    chroma = np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T,axis=0)
    chroma_cqt = librosa.feature.chroma_cqt(y=X, sr=sample_rate)
    mel = np.mean(librosa.feature.melspectrogram(X, sr=sample_rate).T,axis=0)
    contrast = np.mean(librosa.feature.spectral_contrast(S=stft, sr=sample_rate).T,axis=0)
    tonnetz = np.mean(librosa.feature.tonnetz(y=librosa.effects.harmonic(X),sr=sample_rate).T,axis=0)
    chroma_cens = librosa.feature.chroma_cens(y=X, sr=sample_rate)
    S, phase = librosa.magphase(librosa.stft(X))
    # #S = librosa.magphase(librosa.stft(X, window=np.ones, center=False))[0]
    rmse = librosa.feature.rmse(S=S)
    if_gram, D = librosa.ifgram(X)
    centroid = librosa.feature.spectral_centroid(S=np.abs(D), freq=if_gram)
    #2 more variariants of flatness (https://librosa.github.io/librosa/generated/librosa.feature.spectral_flatness.html)
    flatness = librosa.feature.spectral_flatness(y=X)
    spec_bw = librosa.feature.spectral_bandwidth(y=X, sr=sample_rate)
    spec_contrast = librosa.feature.spectral_contrast(S=stft, sr=sample_rate)
    rolloff = librosa.feature.spectral_rolloff(y=X, sr=sample_rate)
    zero_cr = librosa.feature.zero_crossing_rate(X)
    file_dict['index'] = file
    file_dict['mfccs'] = mfccs
    file_dict['chroma'] = chroma
    file_dict['chroma_cqt'] = chroma_cqt
    file_dict['mel'] = mel 
    file_dict['contrast']=contrast
    file_dict['tonnetz']=tonnetz
    file_dict['chroma_cens']=chroma_cens
    file_dict['rmse']=rmse
    file_dict['centroid']=centroid
    file_dict['flatness']=flatness
    file_dict['spec_bw']=spec_bw
    file_dict['spec_contrast']=spec_contrast
    file_dict['rolloff']=rolloff
    file_dict['zero_cr']=zero_cr
    file_dict['mean_mfccs'] = np.mean(mfccs)
    file_dict['mean_chroma'] = np.mean(chroma)
    file_dict['mean_chroma_cqt'] = np.mean(chroma_cqt)
    file_dict['mean_mel'] = np.mean(mel)
    file_dict['mean_contrast'] = np.mean(contrast)
    file_dict['mean_tonnetz']=np.mean(tonnetz)    
    file_dict['mean_chroma_cens']=np.mean(chroma_cens)
    file_dict['mean_rmse']=np.mean(rmse)
    file_dict['mean_centroid']=np.mean(centroid)
    file_dict['mean_flatness']=np.mean(flatness)
    file_dict['mean_spec_bw']=np.mean(spec_bw)
    file_dict['mean_spec_contrast']=np.mean(spec_contrast)
    file_dict['mean_rolloff']=np.mean(rolloff)
    file_dict['mean_zero_cr']=np.mean(zero_cr)
    df = df.append(file_dict, ignore_index=True)

with open('missing_files.pickle', 'wb') as to_write:
    pickle.dump(df, to_write)

#also save to csv as a backup    
df.to_csv('missing_files.csv')
df.head()
```
